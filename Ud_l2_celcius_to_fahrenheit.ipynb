```
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense
```

```
celcius_q=np.array([-40,-20,0,20,40,60,80],dtype='float')
farhenheit_q=np.array([-40,-4,32,68,104,140,176],dtype='float')
```

```
l0=tf.keras.layers.Dense(units=1,input_shape=[1])
```

```
model=tf.keras.Sequential([l0])
```

```
model.compile(optimizer=tf.keras.optimizers.Adam( 0.1), loss='mean_squared_error')
```

```
history=model.fit(x=celcius_q,y=farhenheit_q, epochs=500)
```

```
from matplotlib import pyplot as plt
plt.xlabel='epoch'
plt.ylabel='loss'
plt.plot(history.history['loss'])
```

```
model.predict([200])
```

```
model.get_weights()
```

```
l1=Dense(units=2)
l2=Dense(units=2)
l3=Dense(units=1)
```

```
model1=tf.keras.Sequential([l0,l1,l2,l3])
```

```
model1.compile(optimizer=tf.keras.optimizers.Adam(0.1),loss='mean_squared_error')
```

```
history=model1.fit(x=celcius_q,y=farhenheit_q,epochs=500)
```

```
plt.plot(history.history['loss'])
```

```
model1.predict([200])
```
